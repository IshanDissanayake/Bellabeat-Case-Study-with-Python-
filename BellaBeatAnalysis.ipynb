```python
#########------Average Steps by Id-------###########
dailyActivity['Id'].unique()
dailyActivity['Id'] = dailyActivity['Id'].astype(str)
dailyActivity['DayOfWeek'] = pd.to_datetime(dailyActivity['ActivityDate']).dt.day_name()
avgSteps = dailyActivity.groupby('Id').agg({'TotalSteps':'mean'}).reset_index()  #AVG steps by Id
plt.bar(avgSteps['Id'], avgSteps['TotalSteps'])
plt.xlabel('Id')
plt.ylabel('Avg Total Steps')
plt.title('Average Steps by User')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
#plt.bar(dailyActivity['Id'],dailyActivity['TotalSteps'], color='skyblue')
```

```python
dailyActivity.groupby('DayOfWeek')
```

```python
########---------Average Calories by Id-----------#############
avgCalories = dailyActivity.groupby('Id').agg({'Calories':'mean'}).reset_index()
plt.bar(avgCalories['Id'],avgCalories['Calories'])
plt.xlabel('Id')
plt.ylabel('Avg Calories burnt')
plt.title('Averge Calories Burnt per day by Users')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
###########---------Average time device worn y users----------############
avgDeviceWorn = dailyActivity.groupby('Id').agg({'NoOfMinsDeviceWorn':'mean'}).reset_index()
plt.bar(avgDeviceWorn['Id'],avgDeviceWorn['NoOfMinsDeviceWorn'])
plt.title('Averge Minutes Device Worn per Day by User\n(No of minutes per day = 1440)')
plt.xlabel('Id')
plt.ylabel('Average Minutes Device Worn')
plt.grid(axis='y')
plt.xticks(rotation=90)
plt.show()
```

```python
#############-----------Average Activity level by User-------------###############
avgActivityLevel = dailyActivity.groupby('Id').agg({'VeryActiveMinutes':'mean','FairlyActiveMinutes':'mean','LightlyActiveMinutes':'mean'})
#plt.bar(avgActivityLevel['Id'],avgActivityLevel['VeryActiveMinutes'],avgActivityLevel['FairlyActiveMinutes'],avgActivityLevel['LightlyActiveMinutes'],avgActivityLevel['SedentaryMinutes'])
bar_width = 0.3
spacing = 0.8

# Calculate the positions of each group of bars on the x-axis
x = np.arange(len(avgActivityLevel))

fig, ax = plt.subplots(figsize=(10, 6))
plt.bar(x - bar_width, avgActivityLevel['VeryActiveMinutes'], bar_width, label='VeryActiveMinutes')
plt.bar(x, avgActivityLevel['FairlyActiveMinutes'], bar_width, label='FairlyActiveMinutes')
plt.bar(x + bar_width, avgActivityLevel['LightlyActiveMinutes'], bar_width, label='LightlyActiveMinutes')

plt.xlabel('Users')
plt.ylabel('Average Activity Minutes')
plt.title('Average Activity Level by User')
plt.xticks(x, avgActivityLevel.index)  # Set x-axis labels as user IDs
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
avgActivityLevelByDay = dailyActivity.groupby('DayOfWeek').agg({'VeryActiveMinutes':'mean','FairlyActiveMinutes':'mean','LightlyActiveMinutes':'mean'}).reset_index()
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avgActivityLevelByDay['DayOfWeek'] = pd.Categorical(avgActivityLevelByDay['DayOfWeek'], categories=cats, ordered=True)
avgActivityLevelByDay = avgActivityLevelByDay.sort_values('DayOfWeek')

avgActivityLevelByDay.head()
```

```python
#############-----------Average Activity level by Day of the Week-------------###############
avgActivityLevelByDay = dailyActivity.groupby('DayOfWeek').agg({'VeryActiveMinutes':'mean','FairlyActiveMinutes':'mean','LightlyActiveMinutes':'mean'}).reset_index()
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avgActivityLevelByDay['DayOfWeek'] = pd.Categorical(avgActivityLevelByDay['DayOfWeek'], categories=cats, ordered=True)
avgActivityLevelByDay = avgActivityLevelByDay.sort_values('DayOfWeek')

bar_width = 0.3
spacing = 0.8

# Calculate the positions of each group of bars on the x-axis
x = np.arange(len(avgActivityLevelByDay))

fig, ax = plt.subplots(figsize=(10, 6))
plt.bar(x - bar_width, avgActivityLevelByDay['VeryActiveMinutes'], bar_width, label='VeryActiveMinutes')
plt.bar(x, avgActivityLevelByDay['FairlyActiveMinutes'], bar_width, label='FairlyActiveMinutes')
plt.bar(x + bar_width, avgActivityLevelByDay['LightlyActiveMinutes'], bar_width, label='LightlyActiveMinutes')

plt.xlabel('Day of the Week')
plt.ylabel('Average Activity Minutes')
plt.title('Average Activity Level by Day of the Week')
plt.xticks(x, avgActivityLevelByDay['DayOfWeek'], rotation=90)  # Set x-axis labels as days of the week
plt.grid(axis='y')
plt.legend()


plt.tight_layout()
plt.show()
```

```python
##############-----Average Calories Burnt by Day of the Week-----###############
avgCaloriesByDay = dailyActivity.groupby('DayOfWeek').agg({'Calories':'mean'}).reset_index()
cats = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
avgCaloriesByDay['DayOfWeek'] = pd.Categorical(avgCaloriesByDay['DayOfWeek'], categories=cats, ordered=True)
avgCaloriesByDay = avgCaloriesByDay.sort_values('DayOfWeek')

bar_width = 0.4
spacing = 0.8

x = np.arange(len(avgCaloriesByDay))

fig, ax = plt.subplots(figsize=(4,3))
plt.bar(avgCaloriesByDay['DayOfWeek'],avgCaloriesByDay['Calories'])
plt.xlabel('Day of Week')
plt.ylabel('Average Calories Burnt')
plt.title('Average Calories Burnt by Day of the Week')
plt.xticks(x, avgCaloriesByDay['DayOfWeek'], rotation=90)
plt.grid(axis='y')

plt.tight_layout()
plt.show()

#avgCaloriesByDay['Calories'].max()
```

```python
############-----Heart Rate variation on Days of Week-----#############
heartrateVariationByDay = heartrate_seconds.groupby('DayOfWeek').agg({'Value':'mean'}).reset_index()
cats = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
heartrateVariationByDay['DayOfWeek'] = pd.Categorical(heartrateVariationByDay['DayOfWeek'], categories=cats, ordered=True)
heartrateVariationByDay = heartrateVariationByDay.sort_values('DayOfWeek')

bar_width = 0.4
spacing = 0.8

x = np.arange(len(heartrateVariationByDay))

fig, ax = plt.subplots(figsize=(4,3))
plt.bar(heartrateVariationByDay['DayOfWeek'],heartrateVariationByDay['Value'])
plt.xlabel('Day of Week')
plt.ylabel('Heart Rate')
plt.title('Average Heart Rate by Day of the Week')
plt.xticks(x, heartrateVariationByDay['DayOfWeek'])
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
########---Heart Rate Variation by Hours of the Day------#######
heartrateVariationByHour = heartrate_seconds.groupby('Hour').agg({'Value':'mean'}).reset_index()
heartrateVariationByHour

bar_width = 0.4
spacing = 0.8


fig, ax = plt.subplots(figsize=(8,4))
plt.bar(heartrateVariationByHour['Hour'],heartrateVariationByHour['Value'])
plt.xlabel('Hour')
plt.ylabel('Heart Rate')
plt.title('Average Heart Rate Variation by Hour of the Day')
plt.xticks(heartrateVariationByHour['Hour'], heartrateVariationByHour['Hour'])
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
